```python
# @title $Model \ \ with \ \ different \ \ Transformations$
import matplotlib.pylab as plt
import statsmodels.formula.api as smf

trforms = ['Standard','log_e', 'log_10', 'log_2', 'Square_Root','Square','Exponential','Reciprocal','Power' ]

for tf in trforms:
    print('\033[1m'+tf + ' transformation'+'\033[0m')
    x_col = 'SortingTime'
    y_col = 'DeliveryTime'

    if(tf=='Standard'):
        tf_xcol = x_col
        model=smf.ols(y_col+'~'+tf_xcol, data=df).fit()
        pred = model.predict(df.SortingTime)

    elif(tf=='log_e'):
        tf_xcol = 'np.log(' + x_col + ')'
        model=smf.ols(y_col+'~'+tf_xcol, data=df).fit()
        pred = model.predict(df.SortingTime)

    elif(tf=='log_10'):
        tf_xcol = 'np.log10(' + x_col + ')'
        model=smf.ols(y_col+'~'+tf_xcol, data=df).fit()
        pred = model.predict(df.SortingTime)

    elif(tf=='log_2'):
        tf_xcol = 'np.log2(' + x_col + ')'
        model=smf.ols(y_col+'~'+tf_xcol, data=df).fit()
        pred = model.predict(df.SortingTime)

    elif(tf=='Square_Root'):
        tf_xcol = 'np.sqrt(' + x_col + ')'
        model=smf.ols(y_col+'~'+tf_xcol, data=df).fit()
        pred = model.predict(df.SortingTime)

    elif(tf=='Square'):
        tf_ycol = 'np.sqrt(' + y_col + ')'
        model=smf.ols(tf_ycol+'~'+x_col, data=df).fit()
        pred = model.predict(df.SortingTime)
        pred = (pred)**2

    elif(tf=='Exponential'):
        tf_ycol = 'np.log(' + y_col + ')'
        model=smf.ols(tf_ycol+'~'+x_col, data=df).fit()
        pred = model.predict(df.SortingTime)
        pred = np.exp(pred)

    elif(tf=='Reciprocal'):
        tf_ycol = 'np.reciprocal(' + y_col + ')'
        model=smf.ols(tf_ycol+'~'+x_col, data=df).fit()
        pred = model.predict(df.SortingTime)
        pred = 1/(pred)
    elif(tf=='Power'):
        tf_xcol = 'np.log(' + x_col + ')'
        tf_ycol = 'np.log(' + y_col + ')'
        model=smf.ols(tf_ycol+'~'+tf_xcol, data=df).fit()
        pred = model.predict(df.SortingTime)
        pred = np.exp(pred)


    # For getting coefficients of the varibles used in equation
    print('\033[1m'+'Model Paraeters:'+'\033[0m')
    print(model.params)
    print()

    # P-values for the variables and R-squared value for prepared model
    print('\033[1m'+'Model Summary:'+'\033[0m')
    print(model.summary())
    print()

    print('\033[1m'+'Confidence Interval:'+'\033[0m')
    print(model.conf_int(0.05))
    print()

    # Predicted values of AT using the model
    # pred = model.predict(wcat.Waist)
    # pred = np.exp(pred)
    print('\033[1m'+'Prediction:'+'\033[0m')
    print(pred)
    print()

    # Visualization of regresion line over the scatter plot of Waist and AT
    plt.scatter(x=df['SortingTime'],y=df['DeliveryTime'],color='red');plt.plot(df['SortingTime'],pred,color='black');plt.xlabel('Sorting Time');plt.ylabel('Delivery Time')
    plt.show()
    print()

    #correlation coefficient between the predicted values of AT (`pred`) and the actual values of AT (`wcat.AT`)
    print('\033[1m'+'Correlation between Predicted and Actual Values:'+'\033[0m')
    print(pred.corr(df.DeliveryTime)) # 0.81
    print()
    # Residual Plot
    plt.scatter(x=df.DeliveryTime, y = np.round(df.DeliveryTime - pred,4))
    plt.show()
    print('=====================================================================')
    print()
```

###Conclusion for Que.1: Exponential and Power models are better as both are having higher R-squared scores.
