```python
# @title $Train \ \ Test \ \ Split$

X = df.drop(columns = ["animal name", "type"])
y = df["type"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=108)

print("Shapes")
print("X_train :", X_train.shape)
print("X_test  :", X_test.shape)
print("y_train :", y_train.shape)
print("y_test  :", y_test.shape)
```

```python
# @title $KNN-Classifier \ \ Model$

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train,y_train)
ypred = knn.predict(X_test)

print("Training Accuracy", knn.score(X_train,y_train))
print("Testing Accuracy", knn.score(X_test,y_test))
print()
print(classification_report(y_test,ypred))
```

```python
# @title $Hyperparameter \ \ Tuning \ \ for \ \ KNN$
from matplotlib.pyplot import figure

figure(figsize=(18, 6), dpi=160)

train_acc = []
test_acc = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    train_acc.append(knn.score(X_train,y_train))
    test_acc.append(knn.score(X_test,y_test))

plt.plot(range(1,40), train_acc, color = 'blue', label = 'Training Accuracy')
plt.plot(range(1,40), test_acc, color = 'red', label = 'Testing Accuracy')
plt.vlines(x=3, ymin=0, ymax=1.1, colors = 'Black', linestyles='--')
plt.vlines(x=4, ymin=0, ymax=1.1, colors = 'Black', linestyles='--')
plt.xlabel('K-values')
plt.ylabel('Accuracy')
plt.title('K-value Selection', size = 25)
plt.legend()
```

* We will choose n_neighbor = 3

```python
# @title $Final \ \ KNN-Classifier \ \ Model$

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train,y_train)
ypred = knn.predict(X_test)

print("Training Accuracy", knn.score(X_train,y_train))
print("Testing Accuracy", knn.score(X_test,y_test))
print()
print(classification_report(y_test,ypred))
```

# $Perform \ \ PCA \ \ for \ \ Visualization$

```python
# @title $PCA$
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

sc = StandardScaler()
x = sc.fit_transform(X)

pca = PCA()
pcs = pca.fit_transform(x)
```

```python
# @title $2D-Visualization$
pca = PCA(n_components=2)
pcs = pca.fit_transform(x)

data = pd.DataFrame(pcs,columns=['PC1','PC2'])
data['Class'] = y

figure(figsize=(18, 6), dpi=160)
sns.scatterplot(x = data['PC1'], y = data['PC2'],s = 50, edgecolor = 'black',hue=data['Class'], palette='Set1')
plt.show()
```

```python
# @title $3D-Visualization$

import plotly.express as px

pca = PCA(n_components=3)
pcs = pca.fit_transform(x)
data = pd.DataFrame(pcs,columns=['PC1','PC2','PC3'])
data['Class'] = y

fig = px.scatter_3d(data, x='PC1', y='PC2', z='PC3', color='Class')
fig.show()
```
